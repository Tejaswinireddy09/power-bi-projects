```python
import matplotlib.pyplot as plt
import seaborn as sns
# 1. Bar Plot: GWP Planned vs. GWP
plt.figure(figsize=(12, 6))
sns.barplot(x='Broker Name', y='Planned GWP', data=df1, label='Planned GWP')
sns.barplot(x='Broker Name', y='GWP', data=df1, label='GWP')
plt.xticks(rotation=90)
plt.title('GWP Planned vs. GWP by Broker')
plt.legend()
plt.show()
```

# Top 10 brokers  based on GWP and  type of market 

```python
import pandas as pd

# Sort the data by 'GWP' in descending order
sorted_data = df1.sort_values(by='GWP', ascending=False)

# Split into 'Open Market' and 'Facilities' DataFrames
open_market = sorted_data[sorted_data['Market Type'] == 'Open Market']
facilities = sorted_data[sorted_data['Market Type'] == 'Facilities']

# Get top 10 rows for each market type
top_10_open_market = open_market.head(10)
top_10_facilities = facilities.head(10)

print("Top 10 Brokers for Open Market:")
print(top_10_open_market)
print("\nTop 10 Brokers for Facilities:")
print(top_10_facilities)
```

# CLASS STATS

```python
df2 = pd.read_excel(r"C:\Users\Admin\Downloads\20240209 Sample Dashboard data.xlsx",sheet_name="Class stats")
df2
```

# 1. Stacked Bar Plot: Earned Premium and GWP by Business Class and Year

```python

plt.figure(figsize=(12, 6))
df2.pivot_table(index=['Year', 'Class of Business'], columns='ClassType', values=['Earned Premium', 'GWP '])['Earned Premium'].unstack().plot(kind='bar', stacked=True)
plt.xticks(rotation=45)
plt.title('Earned Premium by Business Class and Year')
plt.xlabel('Year and Business Class')
plt.ylabel('Earned Premium')
plt.legend(title='ClassType')
plt.show()
```

# 2. Line Plot: Planned Business Plan, Earned Premium, and GWP by Business Class and Year

```python

plt.figure(figsize=(12, 6))
for class_of_business in df2['Class of Business'].unique():
    data_class = df2[df2['Class of Business'] == class_of_business]
    plt.plot(data_class['Year'], data_class['Business Plan'], label=f"{class_of_business} Business Plan")
    plt.plot(data_class['Year'], data_class['Earned Premium'], label=f"{class_of_business} Earned Premium")
    plt.plot(data_class['Year'], data_class['GWP '], label=f"{class_of_business} GWP")
plt.title('Planned Business Plan, Earned Premium, and GWP by Business Class and Year')
plt.xlabel('Year')
plt.ylabel('Value')
plt.legend()
plt.show()
```
